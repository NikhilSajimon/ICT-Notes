```python
# The numpy.eye() function in Python's NumPy library creates a two-dimensional array (matrix) with ones on a specified diagonal and zeros elsewhere.
# This function is commonly used to generate identity matrices, which are square matrices with ones on the main diagonal and zeros everywhere else.
a3 = np.eye(4)
a3
```

```python
# Constructing Diagonal Arrays:
# When provided with a 1-dimensional array, numpy.diag() constructs a 2-dimensional array (matrix)
# where the elements of the input 1-D array are placed on the specified diagonal,
# and all other elements are set to zero.

a4 = np.diag([100, 200, 300, 50])
a4
```

```python
# The numpy.arange() function is a fundamental tool within the NumPy library in Python,
# used for creating arrays with evenly spaced values within a specified interval.
# It functions similarly to Python's built-in range() function but returns a NumPy array,
# which is optimized for numerical computations and integrates seamlessly with other NumPy operations.
# syntax -  numpy.arange([start, ]stop, [step, ], dtype=None)

a5 = np.arange(10, 25, 2, dtype=np.float64)
a5
```

```python
# The numpy.reshape() function in Python's NumPy library is used to change the shape of an array without altering its data.
# This function returns a new array (often a view of the original data if possible) with the specified new shape,
# provided the total number of elements remains the same.
# syntax = array.reshape(new_shape) or numpy.reshape(array, new_shape)

np.reshape(a5, (4,2))
```

```python
a5.reshape(2, 4)
```

```python
z = np.random.randint(1, 100, (4,4))
z
```

```python
# np.sum()
# Calculates the sum of all array elements or along a specified axis.
# syntax - np.sum(arr, axis=0) axis= (0 for column, 1 for row)

np.sum(z, 1)
```

```python
np.sum(z, 0)
```

```python
np.sum(z)
```

```python
# np.where()
# The np.where() function is a powerful and very useful NumPy function
# that acts like a vectorized, or element-wise, if-else statement for NumPy arrays.

# Used for data cleaning and handling missing values based on conditions.

# Returns elements chosen from two arrays based on a condition. Useful for conditional assignment.
# np.where(condition, x, y)

np.where(z > 50, z, 0)
```

```python
# np.corrcoef(x, y)
# Calculates the Pearson product-moment correlation coefficients (measures the linear relationship between two datasets).
# np.corrcoef(x, y)

u = np.random.randint(1, 100, (1,10))
v = np.random.randint(1, 100, (1,10))

np.corrcoef(u, v)
```

```python
# np.percentile(a, q, axis=None)
# Computes the q-th percentile of the data along the specified axis.
# Useful for finding quartiles (e.g., q=25,50,75).
# np.percentile(data, 50)

print(u)
print(np.percentile(u, 25))
print(np.percentile(u, 50))
print(np.percentile(u, 75))
```

```python
# np.cumsum(a, axis=None)
# Calculates the cumulative sum of the elements along a given axis.
# np.cumsum(data)

print(z)
print(np.cumsum(z, 0))
print(np.cumsum(z, 1))
```

```python
# np.std(a, axis=None)
# Computes the standard deviation, which measures the amount of variation or dispersion from the mean.
# np.std(data)

np.std(z, 1)
```

```python
np.std(z, 0)
```

```python
# np.var(a, axis=None)
# Computes the variance, which is the average of the squared differences from the mean (standard deviation squared).
# np.var(data)

np.var(z, 0)
```

```python
np.var(z, 1)
```

```python
# np.mean(a, axis=None)
# Calculates the arithmetic mean (average) of the array elements.
# You can specify the axis to get the mean across rows or columns.
# np.mean(data)


# np.median(a, axis=None)
# Computes the median (the middle value of the sorted data).
# np.median(data)


# np.average(a, axis=None, weights=None)
# Computes the weighted average. Very useful when different data points have different levels of importance.
# np.average(data, weights=w)

print(np.mean(z, 0))
print(np.mean(z, 1))
print(np.mean(z))

print(np.median(z, 0))
print(np.median(z, 1))
print(np.median(z))
```

##Pandas

```python
# Core Usages of Pandas

# Data Modeling and Storage:
# Pandas provides the DataFrame structure, which is the standard way to represent and
# store tabular data (like a spreadsheet or SQL table) in Python, giving data meaning
# through labeled rows and columns.

# Data Loading and Export:
# It acts as the primary tool for moving data in and out of Python.
# Pandas can seamlessly read data from various file types (CSV, Excel, SQL databases, JSON) and
# write processed data back out to the same formats.

# Data Cleaning:
# Pandas facilitates rapid data quality management. This includes identifying and
# handling missing values (e.g., ignoring or filling them), removing duplicate entries,
# and correcting inconsistencies across large datasets.

# Data Slicing and Filtering:
# It allows for highly intuitive and powerful ways to select subsets of data.
# Users can easily filter rows based on complex conditions (e.g., all customers over 50 years old in a specific city)
#  or select specific columns by name.

# Aggregation and Summarization:
# Pandas is built for summarizing data.
# The grouping capability allows for quick calculations of metrics (like totals, counts, or averages)
# across different categories (e.g., finding the average sales per region).

# Time Series Analysis:
# It has robust native capabilities for working with data indexed by time,
# making it the go-to tool for tasks like handling irregular time intervals,
# generating date ranges, and preparing financial or sensor data for analysis.
```
